# Difference plot: Nemo Centeral - Nemo South 
### Nemo South does not have drift corrected values; so I am not sure if this is as valuable... but I do not see these "Pauses" in the older data. 

```python
test = pd.merge(df_nemoC, df_nemoS, how ='outer', indicator=True, left_index=True, right_index=True, suffixes=('_C', '_S'))
```

```python
df_nemoMerge = test[test['_merge'] == 'both']
del df_nemoMerge['_merge']
```

```python
df_nemoMerge.head()
```

#### Calculate Depth difference 

```python
depthDiff = df_nemoMerge['RawDep'].values - df_nemoMerge['Depth'].values
```

```python
df_nemoMerge['diff'] = depthDiff
```

```python
df_nemoMerge.hvplot.scatter(x='Date', y='diff', datashade=True, flip_yaxis=True)
```
